# Top 10 Country

```python
data.head()
```

```python
nr_medal = data.groupby('Country')['Medal'].value_counts().unstack(fill_value = 0)
```

```python
nr_medal
    
    
```

```python
nr_medal['TotalMedal'] = nr_medal.sum(axis=1)
```

```python
nr_medal
```

```python
nr_medal.sort_values(by='TotalMedal', ascending=False, inplace=True)
```

```python
top_10_country = nr_medal.head(10)
```

```python
top_10_country
```

```python
plt.bar(top_10_country.index , top_10_country['TotalMedal'])
plt.xlabel("Country")
plt.ylabel('Nr Of Medals')
plt.title('Top 10 Country')
plt.xticks(rotation=45, ha='right')
```

# Evolution of medal numbers in comparation with year

```python
data.head()
```

```python
evolution_per_year = data.groupby('Year')['Medal'].count()
evolution_per_year
```

```python
plt.figure(figsize=(10, 6))
evolution_per_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Nr Total Medal')
plt.title('Evolution')
plt.grid(True)
plt.tight_layout()
```

# % of Male & Female 

```python
data.head()
```

```python
nr_men = (data['Gender']=='Men').sum()
nr_men
```

```python
nr_women = (data['Gender'] == 'Women').sum()
nr_women
```

```python
total_athlete = len(data)
```

```python
perc_men = (nr_men / total_athlete ) * 100 
perc_women = (nr_women / total_athlete) * 100
```

```python
plt.bar(['Men','Women'],[perc_men , perc_women])
plt.xlabel('Gender')
plt.ylabel('Percent')
```
